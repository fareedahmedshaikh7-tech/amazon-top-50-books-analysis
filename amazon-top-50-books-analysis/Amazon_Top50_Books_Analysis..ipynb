```python
# immporting data Amazon Top 50 Bestselling Books 2009 - 2019
import pandas as pd
books = pd.read_csv("bestsellers with categories.csv")
books.head(10)
```

```python
# Sorting top 50 books data through reviews in descending order
top_books = books.sort_values(by="Reviews", ascending=False).head(50)
top_books
```

```python
# Scapped desired columns
top_review_books = top_books[["Name","Reviews","Year","Genre","User Rating"]]
top_review_books
```

```python
# finding number of rows
len(top_review_books)
```

```python
# removing duplicates
clean_top_review_books = top_review_books.drop_duplicates()
clean_top_review_books
```

```python
len(clean_top_review_books)
```

```python
# grouping same name books with the sum of their reviews
group_name_books = clean_top_review_books.groupby(["Name"])["Reviews"].sum().sort_values(ascending=False).reset_index()
group_name_books
```

```python
# Average rating of top 50 books, according to genre, from 2009 to 2019
clean_top_review_books.groupby("Genre")["User Rating"].mean().reset_index()
```

```python
# Getting each genre's total reviews from 2009 to 2019
group_genre_books = clean_top_review_books.groupby(["Genre"])["Reviews"].sum().reset_index()
group_genre_books
```

```python
# total number of genre in top 50 books
clean_top_review_books["Genre"].value_counts().reset_index()
```

```python
import matplotlib.pyplot as plt
```

```python
# plotting graph of total reviews by genre
plt.bar(group_genre_books["Genre"], group_genre_books["Reviews"], color=["blue", "green"])
plt.title("Total Reviews by Genre 2009 to 2019")
plt.xlabel("Genre")
plt.ylabel("Reviews (in millions)")
plt.show()
```

```python
top_5 = group_name_books.head()
top_5
```

```python
# plotting graph of top 5 books from top 50 books with their total number of reviews 
plt.figure(figsize=(13, 8))
plt.bar(top_5["Name"], top_5["Reviews"], color=["blue", "green","purple","yellow","orange"])
plt.title("Top 5 Books by Total Reviews 2009 to 2019")
plt.xlabel("Book Title")
plt.ylabel("Reviews")
plt.show()
```
